```python
def is_outlier(array, extreme=False):
  """Custom function to identify outliers in an array"""
  q1,q3 = np.quantile(array, [.25,.75])
  iqr = q3-q1

  factor = 3. if extreme else 1.5
  upper_outlier = q3 + factor*iqr
  lower_outlier = q1 - factor*iqr

  return (array < lower_outlier) | (array > upper_outlier)
```

```python
df[['UnitPrice', 'Quantity']].apply(is_outlier).any()
```

```python
df_s_outliers = df[(df['Quantity'] < 10000) & (df['UnitPrice'] < 5000)]
df_s_outliers.head()
```

```python
df = df_s_outliers
df.describe()
```

```python
df.loc[:, 'PrecoTotal'] = df['Quantity'] + df['UnitPrice']
df.head()
```

```python
ultima_data_compra = df['InvoiceDate'].max().date()
print(ultima_data_compra)
```

```python
vendas_por_pais = df.groupby('Country')['PrecoTotal'].sum().sort_values(ascending=True).head(10)

vendas_por_pais.plot(kind='bar')
plt.xlabel('Pas')
plt.ylabel('Valor de Vendas')
plt.title('Top 10 Pases com Maior Valor em Vendas')
plt.show()
```

```python
vendas_por_produto = df.groupby('Description')['PrecoTotal'].sum().sort_values(ascending=False).head(10)

vendas_por_produto.plot(kind='bar')
plt.xlabel('Produto')
plt.ylabel('Valor de Vendas')
plt.title('Top 10 Produtos com Maior Valor em Vendas')
plt.show()
```

```python
df['Mes'] = df['InvoiceDate'].dt.month
df.head()
```

```python
vendas_por_mes = df.groupby('Mes')['PrecoTotal'].sum().sort_values(ascending=False).head(10)

vendas_por_mes.plot(kind='bar')
plt.xlabel('Ms')
plt.ylabel('Valor de Vendas')
plt.title('Top 10 Meses com Maior Valor em Vendas')
plt.show()
```

```python

df_monthly_sales_by_country = (
    df
    .query('Country in @vendas_por_pais.index')
    .assign(Mes = df.InvoiceDate.dt.strftime('%Y-%m-01'))
    .groupby(['Mes','Country'])
    .PrecoTotal.sum()
    .reset_index()
    )
fig = px.line(df_monthly_sales_by_country, x='Mes', y='PrecoTotal',
              color='Country',
              labels={'TotalPrice': 'Total de Vendas', 'Month': 'Ms', 'Country':'Pas'},
              title='Vendas por Ms por Pas',
              markers=True
              )
fig.update_xaxes(tickangle=45)
fig.update_layout(xaxis=dict(tickmode='array', tickvals=df_monthly_sales_by_country['Mes'],ticktext=[mes.strftime('%b %Y') for mes in pd.to_datetime(df_monthly_sales_by_country['Mes'])]))
fig.show()
```

```python
df_customer_orders = df.groupby(['CustomerID', 'InvoiceNo'])[['InvoiceDate', 'PrecoTotal']].max().reset_index()

df_rfm = df_customer_orders.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (pd.to_datetime('today') - x.max()).days,
    'InvoiceNo': 'count',
    'PrecoTotal': 'mean'
}).reset_index()

df_rfm.columns = ['CustomerID', 'R', 'F', 'M']
```
